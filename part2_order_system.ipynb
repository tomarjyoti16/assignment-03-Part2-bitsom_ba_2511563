Cart Operations

```python
cart = []

def add_to_cart(item, quantity):
  if item not in menu :
    print(f"Item not in menu")
    return

  if not menu[item]["available"]:
    print(f"Item not available currently")
    return

  for cart_item in cart:
    if cart_item["item"] == item:
      cart_item["quantity"] += quantity
      print (f"{quantity} {item} added to cart")
      print(f"Cart: {cart}")
      return

  cart.append({
      "item": item,
      "quantity": quantity,
      "price": menu[item]["price"]
      })
  print(f"{quantity} {item} added to cart")

  print(f"Cart: {cart}")
```

Remove from cart

```python
def remove_from_cart(item):
  item_found_and_removed = False
  for cart_item in cart:
    if cart_item["item"] == item:
      cart.remove(cart_item)
      print(f"{item} removed from cart")
      print(f"Cart: {cart}")
      item_found_and_removed = True
      break

  if not item_found_and_removed:
    print(f"{item} not found in cart")
```

Update the quantity

```python
def update_quantity(item, new_quantity):
  item_found_and_updated = False
  for cart_item in cart:
    if cart_item["item"] == item:
      cart_item["quantity"] = new_quantity
      print(f"Quantity of {item} updated to {new_quantity}")
      print(f"Cart: {cart}")
      item_found_and_updated = True
      break

  if not item_found_and_updated:
    print(f"{item} not found in cart")
```

```python
#frinal order summary creation
def order_summary():
  subtotal = 0

  print(f"============Order Summary============")

  for cart_item in cart:
    item_price = cart_item["quantity"] * cart_item["price"]
    subtotal += item_price

    print(f"{cart_item["item"]:<20} x{cart_item["quantity"]}   ₹{item_price:.2f}")

  print(f"-------------------------------------")

  gst = subtotal * 0.05
  total = subtotal + gst
#print bill details



  print(f"{'Subtotal:':<20}  ₹{subtotal:.2f}")
  print(f"{'GST (5%):':<20}  ₹{gst:.2f}")
  print(f"-------------------------------------")
  print(f"{'Total:':<20}  ₹{total:.2f}")
  print(f"====================================")


```

## **Task 3 — Inventory Tracker with Deep Copy**

```python
import copy

#Deep Copy
inventory_backup = copy.deepcopy(inventory)
#Modify Original inventory
inventory["Paneer Tikka"]["stock"] = 15
#print both quantities
print(f"Modified stock:")
print(inventory["Paneer Tikka"]["stock"])
print(f"Backup stock:")
print(inventory_backup["Paneer Tikka"]["stock"])

#Restore Inventory
inventory = copy.deepcopy(inventory_backup)

#Print restored inventory
print(f"Restored stock:")
print(inventory["Paneer Tikka"]["stock"])

print(f"Whole inventory:")
print(inventory)


```

Order Fulfillment

```python
def order_fulfill():
  for cart_item in cart:
    name = cart_item["item"]
    quantity = cart_item["quantity"]

    stock_available = inventory[name]["stock"]

    if stock_available >= quantity:
      inventory[name]["stock"] -= quantity
      print(f"{quantity} {name} fulfilled")
    else:
      print(f"{name} Only {stock_available} available, can't fulfill {quantity}")
      inventory[name]["stock"] = 0 #Deducting what ever available

    #Reorder alert



  for item, details in inventory.items():

    if details["stock"] < details["reorder_level"]:
      print("==========Reorder Alert=============")
      print(f"⚠ Reorder Alert: {item} - Only {details['stock']} unit(s) available (Reorder level: {details["reorder_level"]})")

```

```python
order_fulfill()
```

```python
def Sales_Log_Analysis():

  daily_revenue = {}
  for date, orders in sales_log.items():
    total = sum(order["total"] for order in orders)
    daily_revenue[date] = total

  print(f"Total Revenur per day")
  for date, revenue in daily_revenue.items():
    print(f"{date}: ₹{revenue:.2f}")

  #Finding Best selling day
  best_selling_day = max(daily_revenue, key=daily_revenue.get)
  print(f"Best Selling Day: {best_selling_day} with revenue of , {daily_revenue[best_selling_day]}")

  #Checking most ordered item
  from collections import Counter

  item_counter = Counter()

  for orders in sales_log.values():
    for order in orders:
      item_counter.update(order["items"])

  most_ordered_item = item_counter.most_common(1)[0]
  print(f"Most Ordered Item: {most_ordered_item[0]}, with, {most_ordered_item[1]}, orders")

```

```python
Sales_Log_Analysis()
```
